### Basic array operations

This section covers addition, subtraction, multiplication, division, and more

```python
import numpy as np
```

```python
# Once you’ve created your arrays, you can start to work with them. Let’s say, for example, that you’ve created two arrays,
# one called “data” and one called “ones”

# you can add these arraystogether with a plus sign

data = np.array([1, 2])
ones = np.ones(2, dtype=int)
data + ones
```

```python
# you can perform all basic calculations :
print(data - ones)
print(data * data)
print(data / data)
```

```python
# Basic operations are simple with NumPy. If you want to find the sum of the elements in an array, you’d use sum(). 
# This works for 1D arrays, 2D arrays, and arrays in higher dimensions

a = np.array([1, 2, 3, 4])

a.sum()
```

### In 2d Array

```python
b = np.array([[1, 1], [2, 2]]) # -> [1 1]
                               # -> [2​ 2​]

# You can sum over the axis of rows with:
print(b.sum(axis=0))

# You can sum over the axis of columns with:
print(b.sum(axis=1))

# axis=0 → move down rows (vertical operation)
# axis=1 → move across columns (horizontal operation)
```
